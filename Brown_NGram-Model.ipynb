# BIGRAM

```python
# BIGRAM
import random
from collections import defaultdict

# Sample Text
text_1 = "The quick brown fox jumps over the lazy dog."
text_2 = "Data science is an interdisciplinary field that uses scientific methods."
text_3 = "Data science is fun and informative. Data science is powerful and useful."

# Create a Bigram
n = 2
ngrams = defaultdict(list)
words = text_3.split()

# Loop to create trigrams
for i in range(len(words) - n + 1):
    gram = tuple(words[i:i+n])
    next_word = words[i+n] if i+n < len(words) else None

    # Add next word if it's not None
    if next_word:
        ngrams[gram].append(next_word)

# User-Input
user_input = input("Enter a word or phrase: ")    
input_words = user_input.split()

if len(input_words) != 2:
    print('Please enter a 2-word input.')
else:
    # Find matching bigrams
    matching_grams = [gram for gram in ngrams if gram[-1] == input_words[-1]]
    
    if matching_grams:
        current_gram = random.choice(matching_grams)
        result = list(current_gram)
    
        # Auto-complete Suggestion
        for _ in range(10):
            if current_gram in ngrams and ngrams[current_gram]:
                next_word = random.choice(ngrams[current_gram])
                result.append(next_word)
                current_gram = tuple(result[-n:])
            else:
                break
        
        print("Auto-complete Suggestion: ", " ".join(result))
    else:
        print("No matching bigram found.")
```
